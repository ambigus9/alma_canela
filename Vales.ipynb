```python
def porcent_null(df):
    # Función que determina el porcentaje de registros nulos
    df = (df.isnull().sum()/len(df))*100
    return df[df > 0]
```

```python
def print_full(x):
    pd.set_option('display.max_rows', len(x))
    print(x)
    pd.reset_option('display.max_rows')
```

```python
print_full(porcent_null(pd.read_csv('source/detafave.csv')))
```

## Filtramos Columnas más Relevantes

```python
relevantes = ['CODCLI', 'CODPP', 'CODOPE', 'CANVEN', 'CANOPE', 'COSTOOPE', 'COSTOOPEC', 'TICKET', 'NROTICK', 'FECHATICK', 'HORATICK', 'FECHAITT', 'HORAITT', 'FECHAFTT', 'HORAFTT', 'FECHAIOPE', 'HORAIOPE', 'FECHAFOPE', 'HORAFOPE', 'IDPROTICKETS', 'FECHA', 'HORA']
df2 = df[relevantes]
df2.iloc[:10,:]
```

```python
df = df[relevantes]
df['TIMESTAMP'] = pd.to_datetime(df['FECHA'] + ' ' + df['HORA'])
df = df.set_index('TIMESTAMP')
df
```

```python
def plot_time_line(df):

    data = [go.Scatter(
              x=df.index,
              y=df['COSTOOPE'])]

    return py.iplot(data)
```

## Filtramos por Año

Analizamos la cantidad de transacciones por cada año

```python
plot_time_line(df["2016"].groupby('FECHA').count())
```

```python
plot_time_line(df["2017"].groupby('FECHA').count())
```

```python
df.groupby(pd.TimeGrouper(freq='M')).count()
```

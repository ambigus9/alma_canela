```python
def plot_time_line(df,title,x_axis,y_axis):

    data = [go.Scatter(
              x=df.index,
              y=df['COSTOOPE'])]
    
    layout = go.Layout(
        title=title,
        xaxis=dict(
            title=x_axis,
            titlefont=dict(
                family='Courier New, monospace',
                size=18,
                color='#7f7f7f'
            )
        ),
        yaxis=dict(
            title=y_axis,
            titlefont=dict(
                family='Courier New, monospace',
                size=18,
                color='#7f7f7f'
            )
        )
    )
    fig = go.Figure(data=data, layout=layout)
    return py.iplot(fig, filename=title)
```

```python
def plot_time_line_two(df,df_,title,x_axis,y_axis):

    trace1 = go.Scatter(
        x=df.index,
        y=df['COSTOOPE'],
        name='2016'
    )
    trace2 = go.Scatter(
        x=df_.index,
        y=df_['COSTOOPE'],
        name='2017'
    )
    data = [trace1, trace2]
    
    layout = go.Layout(
        title=title,
        xaxis=dict(
            title=x_axis,
            titlefont=dict(
                family='Courier New, monospace',
                size=18,
                color='#7f7f7f'
            )
        ),
        yaxis=dict(
            title=y_axis,
            titlefont=dict(
                family='Courier New, monospace',
                size=18,
                color='#7f7f7f'
            )
        )
    )
    fig = go.Figure(data=data, layout=layout)
    return py.iplot(fig, filename=title)
```

```python
def plot_two_bar(x,y,y2,t1_nam,t2_nam,x_axis,y_axis,title):

    x = x
    y = y
    y2 = y2

    trace1 = go.Bar(
        x=x,
        y=y,
        name=t1_nam,
        textposition = 'auto',
        marker=dict(
            color='rgb(158,202,225)',
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
            ),
        opacity=0.6
    )

    trace2 = go.Bar(
        x=x,
        y=y2,
        name=t2_nam,
        textposition = 'auto',
        marker=dict(
            color='rgb(58,200,225)',
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
            ),
        opacity=0.6
    )

    data = [trace1,trace2]

    layout = go.Layout(
        title=title,
        xaxis=dict(
            title=x_axis,
            titlefont=dict(
                family='Courier New, monospace',
                size=18,
                color='#7f7f7f'
            )
        ),
        yaxis=dict(
            title=y_axis,
            titlefont=dict(
                family='Courier New, monospace',
                size=18,
                color='#7f7f7f'
            )
        )
    )
    fig = go.Figure(data=data, layout=layout)
    return py.iplot(fig, filename=title)
```

```python
def nombres_meses(df):
    list_temp = []
    # Determinados el mes de cada fecha
    df['MONTHFOPE'] = pd.DataFrame(pd.DatetimeIndex(df['FECHAFOPE'])).FECHAFOPE.dt.month

    for i in df['MONTHFOPE']:
        for j in range(1,len(meses)+1):
            if i==j:
                list_temp.append(meses[j-1])

    df['MONTHFOPE'] = list_temp
    return df
```

```python
mesesitos = [df_1d,df_1d_3d,df_3d_7d,df_7d_10d,df_10d_14d,df_14d]

for i in range(len(mesesitos)):
    nombres_meses(mesesitos[i])
```

```python
def filtro_ano(df,ano):
    return df[(pd.to_datetime(df['FECHAIOPE']) > pd.to_datetime(ano+'-01-01'))&(pd.to_datetime(df['FECHAIOPE']) < pd.to_datetime(str(int(ano)+1)+'-01-01'))]
```

```python
def ticktes_mes(year):
    list_df = []
    for i in range(len(mesesitos)):
        list_df.append(filtro_ano(mesesitos[i],year).groupby(['MONTHFOPE']).sum().reset_index())

    list_df[0]['RANGE'] = '< 1 día'
    list_df[1]['RANGE'] = '1 a 3 días'
    list_df[2]['RANGE'] = '3 a 7 días'
    list_df[3]['RANGE'] = '7 a 10 días'
    list_df[4]['RANGE'] = '10 a 14 días'
    list_df[5]['RANGE'] = '> 14 días'

    return pd.concat([list_df[0],list_df[1],list_df[2],list_df[3],list_df[4],list_df[5]])
```

```python
tickets_performance_2016 = ticktes_mes('2016')
tickets_performance_2017 = ticktes_mes('2017')

tickets_performance_2016.to_csv('tickets_performance_2016')
tickets_performance_2017.to_csv('tickets_performance_2017')
```

```python
def plot_heatmap(df,x_nam,y_nam,title):
    data = [
        go.Heatmap(
            z=df.iloc[:,1],
            x=df.iloc[:,0],
            y=df.iloc[:,2],
            colorscale='Viridis',
        )
    ]

    layout = go.Layout(
        title=title,
        xaxis = dict(title=x_nam),
        yaxis = dict(title=y_nam)
    )

    fig = go.Figure(data=data, layout=layout)
    return py.iplot(fig, filename=title)
```

```python
def mean_time(df):
    df['TIMEDIF'] = df['TIMEDIF'].dt.total_seconds()

    aggregator = {'NROTICKC' : 'sum',
                  'TIMEDIF' : 'mean'}

    df_new = nombres_meses(df).groupby('MONTHFOPE').agg(aggregator)

    df_new['TIMEDIF'] = pd.DataFrame(pd.to_timedelta(df_new.TIMEDIF.astype(int), unit='s'))
    return df_new
```
